```python
sales_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Sales_Transactions.csv")
sales_df
```

```python
products_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Product_Master.csv")
products_df
```

```python
sales_df['Revenue'] = sales_df['Selling_Price'] * sales_df['Quantity']
sales_df
```

```python
total_revenue = sales_df['Revenue'].sum()
print(total_revenue)
```

```python
#Total Spend
total_spend = sales_df.groupby('Customer_ID')['Revenue'].sum()
total_spend
```

```python
# Order Frequency
order_freq = sales_df.groupby('Customer_ID')['Order_ID'].nunique()
order_freq
```

```python
# Avg Order Value
aov = total_spend / order_freq
aov
```

```python
#Return Frequency
```

```python
returns_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Returns.csv")
returns_df
```

```python
merged_df = sales_df.merge(
    returns_df[['SKU_ID', 'Return_ID']],
    on='SKU_ID',
    how='left'
)
merged_df
```

```python
return_freq = merged_df.groupby('Customer_ID')['Return_ID'].count()
return_freq
```

```python
# Combineing all togeather
customer_df = pd.DataFrame({
    'Total_Spend': total_spend,
    'Order_Frequency': order_freq,
    'Avg_Order_Value': aov,
    'Return_Frequency':return_freq
}).reset_index()
customer_df
```

```python
##CLUSTRING
```

```python
from sklearn.preprocessing import StandardScaler

features = ['Total_Spend', 'Order_Frequency', 'Avg_Order_Value', 'Return_Frequency']

scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_df[features])
```

```python
def label_segment(row):
    if row['Total_Spend'] > 10000 and row['Order_Frequency'] > 10:
        return 'High Value Loyal'
    elif row['Avg_Order_Value'] < 500 and row['Order_Frequency'] > 5:
        return 'Deal Hunters'
    elif row['Order_Frequency'] == 1:
        return 'One-time Buyers'
    elif row['Return_Frequency'] > 3:
        return 'High Return Risk'
    else:
        return 'Regular Customers'

customer_df['Segment'] = customer_df.apply(label_segment, axis=1)
```

```python
sales_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Sales_Transactions.csv")
sales_df
```

```python
returns_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Returns.csv")
returns_df
```

```python
products_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Product_Master.csv")
products_df
```

```python
vendor_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Vendor_Supply.csv")
vendor_df
```

```python
inventory_df = pd.read_csv(r"C:\Users\pavan_shsjuzl\Downloads\Inventory_Daily.csv")
inventory_df
```

```python
merged_df = sales_df.merge(
    products_df[['SKU_ID', 'Unit_Cost']],
    on='SKU_ID',
    how='left'
)
merged_df
```

```python
sales_df['Profit'] = (sales_df['Selling_Price'] - merged_df['Unit_Cost']) * sales_df['Quantity']
```

```python
#Top 5 loss Drivers
loss_drivers = (
    sales_df.groupby('SKU_ID')['Profit']
    .sum()
    .sort_values()
    .head(5)
    .reset_index(name='Total_Loss')
)
loss_drivers
```

```python
#Worst 3 Vendors
worst_vendors = (
    vendor_df.groupby('Vendor_ID')['Supply_Delay_Days']
    .mean()
    .sort_values(ascending=False)
    .head(3)
    .reset_index(name='Avg_Delay')
)
worst_vendors
```

```python
#Stockout Hotspots
stockout_hotspots = inventory_df[
    inventory_df['Closing_Stock'] == 0
].groupby('Store_ID').size().reset_index(name='Stockout_Count')

stockout_hotspots = stockout_hotspots.sort_values(
    by='Stockout_Count', ascending=False
).head(5)
stockout_hotspots
```

```python
sales_df['Profit'] = (sales_df['Selling_Price'] - merged_df['Unit_Cost']) * sales_df['Quantity']
```

```python
best_skus = (
    sales_df.groupby('SKU_ID')['Profit']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index(name='Total_Profit')
)
best_skus
```

```python
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.styles import getSampleStyleSheet

doc = SimpleDocTemplate("Executive_Report.pdf")
styles = getSampleStyleSheet()

content = []

# Title
content.append(Paragraph("Retail Insights Executive Report", styles['Title']))
content.append(Spacer(1, 12))

# Top Loss Drivers
content.append(Paragraph("Top 5 Loss Drivers", styles['Heading2']))
for i, row in loss_drivers.iterrows():
    text = f"SKU {row['SKU_ID']} → Loss: {row['Total_Loss']}"
    content.append(Paragraph(text, styles['Normal']))
content.append(Spacer(1, 10))

# Worst Vendors
content.append(Paragraph("Worst 3 Vendors", styles['Heading2']))
for i, row in worst_vendors.iterrows():
    text = f"Vendor {row['Vendor_ID']} → Avg Delay: {row['Avg_Delay']} days"
    content.append(Paragraph(text, styles['Normal']))
content.append(Spacer(1, 10))

# Stockouts
content.append(Paragraph("Stockout Hotspots", styles['Heading2']))
for i, row in stockout_hotspots.iterrows():
    text = f"Store {row['Store_ID']} → Stockouts: {row['Stockout_Count']}"
    content.append(Paragraph(text, styles['Normal']))
content.append(Spacer(1, 10))

# Best SKUs
content.append(Paragraph("Best SKUs to Promote", styles['Heading2']))
for i, row in best_skus.iterrows():
    text = f"SKU {row['SKU_ID']} → Profit: {row['Total_Profit']}"
    content.append(Paragraph(text, styles['Normal']))

doc.build(content)
```
