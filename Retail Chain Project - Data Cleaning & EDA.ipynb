```python
#SALES_TRANSACTIONS - DATA CLEANING AND EDA
```

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_excel(r"C:\Users\pavan_shsjuzl\Downloads\Retail Chain Profit Leakage & Demand Optimization\Data\1.Raw Data\Retail_Profit_Leakage_Dataset.xlsx",
sheet_name="Sales_Transactions")
df
```

```python
df.info
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
#Data Cleaning -(formatting date, removing inappropriate dates(612 rows))
df['Order_Date'] = pd.to_datetime(df['Order_Date'], errors='coerce').dt.date
df = df.dropna(subset=['Order_Date'])
df
```

```python
#Data Cleaning - Formatting and stanadardizing Store_ID
df.loc[:, 'Store_ID'] = (
    df['Store_ID']
    .astype(str)
    .str.upper()
    .str.strip()
    .str.replace('_', '-', regex=False)
    .str.replace(r'([A-Z]+)-?(\d+)', lambda x: f"{x.group(1)}-{int(x.group(2)):02d}", regex=True)
)
df
```

```python
#Data Cleaning - Standardizing Order_ID & SKU_ID by removing "-" between the code
df.loc[:, 'Order_ID'] = df['Order_ID'].astype(str).str.replace('-', '', regex=False)
df.loc[:, 'SKU_ID'] = df['SKU_ID'].astype(str).str.replace('-', '', regex=False)
df
```

```python
# Clean and standardize Promo_ID
clean = df['Promo_ID'].astype(str).str.upper().str.strip()

nums = clean.str.replace(r'[^0-9]', '', regex=True)
nums = nums.str[:3].str.zfill(3)

df.loc[:, 'Promo_ID'] = 'PRO' + nums

df.loc[clean.isin(['', 'NAN', 'NONE']), 'Promo_ID'] = 'Unknown'
df
```

```python
#Replacing Negative Qty values to "0"
df.loc[df['Quantity'] < 0, 'Quantity'] = 0
df
```

```python
#Data Cleaning - (Remove rows where order_status is Cancelled)
df = df[df['Order_Status'] != 'Cancelled'].reset_index(drop=True)
df
```

```python
#Data Cleaning (Removed blank sales values,replace blank discount_amt with "Unknown")
df['Discount_Amount'] = df['Discount_Amount'].fillna(0)
df = df.dropna(subset=['Selling_Price'])
df
```

```python
#Data Cleaning (Fill blank payment_mode with "Unknown")
df.loc[:,'Payment_Mode'] = df['Payment_Mode'].fillna('Unknown')
df
```

```python
df.isna().sum()
```

```python
#Data Cleaning - Remove duplicate Order_ID
df = df.drop_duplicates(subset = ['Order_ID'])
df
```

```python
df.duplicated().sum()
```

```python
df.to_excel("Sales_Transactions_ipynb.xlsx", index=False)
```
